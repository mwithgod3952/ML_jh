##### Dignosis - Performance Evaluation
    - Once the model has been specified and the parameters estimated, the model should be checked for adequacy *** 
    - Use the model to forcast all of the known values of the data and investigate the residuals, and generate the ACF plot for the residuals
$residual : y-\hat{y}$

    - Rule: The model is OK if more than **two or three out of 40 fall outside the bound or if  one falls far outside the bounds.**
